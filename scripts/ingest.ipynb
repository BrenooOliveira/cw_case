```python
import json
from configparser import ConfigParser
import psycopg2
```

## Credential Configs

```python
CONFIG_PATH = "config.ini"
config = ConfigParser()
config.read(CONFIG_PATH)
HOST = config.get("POSTGRES", "host")
PORT = config.get("POSTGRES", "port")
DATABASE = config.get("POSTGRES", "database")
USER = config.get("POSTGRES", "user")
PASSWORD = config.get("POSTGRES", "password")

```

## DB Conection

```python
conn = psycopg2.connect(
    host=HOST,
    database=DATABASE,
    user=USER,
    password=PASSWORD,
    port=PORT
)
```

```python
cur = conn.cursor()
```

## Create Table

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS kyc_data (
    id SERIAL PRIMARY KEY,
    creation_date TIMESTAMP NOT NULL DEFAULT NOW(),
    client_data JSONB NOT NULL
);
""")

conn.commit()

```

```python
with open("/home/breno-oliveira/Documentos/gitRepositories/cw_case/data/sarah_data.json") as f:
    data = json.load(f)

cur.execute(
    "INSERT INTO kyc_data (client_data) VALUES (%s)",
    [json.dumps(data)]
)

conn.commit()
```

## Simple Query to check data insertion

```python
cur.execute("SELECT * FROM kyc_data LIMIT 10;")
rows = cur.fetchall()
for row in rows:
    print(row)
```

```python
cur.close()
conn.close()
```
